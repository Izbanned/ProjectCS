5.1 Combine all csv files into one

```python
import pandas as pd
import os
from tqdm import tqdm

detail = []
folder_path = 'dataset' 
csv_files = os.listdir(folder_path)

for file in tqdm(csv_files):
    try:
        file_path = os.path.join(folder_path, file)
        df = pd.read_csv(file_path, na_values="-", low_memory=False)
        df["Name"] = file[:-4]  
        df["busNumber"] = file[1:4]  
        df["time_iso"] = pd.to_datetime(df["time_iso"])  
        detail.append(df)
    except ValueError as e:
        print(f"Skipping file {file} due to error: {e}")

detail = pd.concat(detail)  
```

## **Task 15.** *For each bus compute the total time when the halt brake is active and the total time when the park brake is active. Compute also the ratio between those two times.*

```python
def compute_brake_times(detail):
    detail["time_diff"] = detail.groupby("Name")['time_iso'].diff().fillna(pd.Timedelta(seconds=1)).dt.total_seconds()

    def compute_total_active_time(brake_status):
        return (detail[detail[brake_status]==1]
                .groupby("Name")["time_diff"]
                .sum()
                .reset_index()
                .rename(columns={"Name": "Name", "time_diff": f"Total_{brake_status}_Time"}))

    halt_brake_active_time = compute_total_active_time('status_haltBrakeIsActive')
    park_brake_active_time = compute_total_active_time('status_parkBrakeIsActive')

    brake_times = pd.merge(halt_brake_active_time, park_brake_active_time, on="Name")
    brake_times["Brake_Time_Ratio"] = brake_times["Total_status_haltBrakeIsActive_Time"] / brake_times["Total_status_parkBrakeIsActive_Time"]

    return brake_times

brake_times = compute_brake_times(detail)
brake_times
```

## **Task 16.** *For each pair of stops that are consecutive in at least a trip, compute the average speed achieved when going from the first to the second stop.*

```python
def compute_trip_data(trip_rows):
    df = pd.DataFrame()
    df["consecutive_stops"] = trip_rows["itcs_stopName"] + " TO " + trip_rows["itcs_stopName"].shift(-1)
    df["average_speed"] = (trip_rows["odometry_vehicleSpeed"] + trip_rows["odometry_vehicleSpeed"].shift(-1)) / 2
    return df.dropna()

valid_rows = detail[detail["itcs_stopName"].notna()]

trip_dataframes = valid_rows.groupby("Name").apply(compute_trip_data)

average_speeds = trip_dataframes.groupby("consecutive_stops")["average_speed"].mean().reset_index()

average_speeds
```
